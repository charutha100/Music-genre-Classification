```python
df2=pd.DataFrame(df2,columns=['Popularity', 'danceability', 'energy',
       'key', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_in min/ms'])
df2
```

```python
df=df.drop(['Popularity', 'danceability', 'energy',
       'key', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_in min/ms'],axis='columns')
df
```

```python
df=pd.concat([df,df2],axis='columns')
df
```

```python
df=df.drop(['Artist Name', 'Track Name'],axis='columns')
df
```

```python
Y=df['Class']
X=df.drop(['Class'],axis='columns')
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=77)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(multi_class='multinomial')
lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

STANDARD VECTOR CLASSIFIERS

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
svc.score(X_test,Y_test)
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
rf.score(X_test,Y_test)
```

K NEIGHBOURSS CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
kn.fit(X_train,Y_train)
```

```python
kn.score(X_test,Y_test)
```

XGBOOST CLASSIFIER

```python
from xgboost import XGBClassifier
xg=XGBClassifier()
xg.fit(X_train,Y_train)
```

```python
xg.score(X_test,Y_test)
```

NAIVE BAYES CLASSIFIER

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
gnb.score(X_test,Y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(lr,X,Y,cv=10)
np.average(scores)
```

```python
scores=cross_val_score(svc,X,Y,cv=10)
np.average(scores)
```

```python
scores=cross_val_score(rf,X,Y,cv=10)
np.average(scores)
```

```python
scores=cross_val_score(kn,X,Y,cv=10)
np.average(scores)
```

```python
scores=cross_val_score(xg,X,Y,cv=10)
np.average(scores)
```

```python
scores=cross_val_score(gnb,X,Y,cv=10)
np.average(scores)
```

EVALUATION OF METRICS

```python
models=[lr,svc,rf,kn,xg,gnb]
accuracies=[]
for i in models:
    scores=cross_val_score(i,X,Y,cv=10)
    accuracies.append(np.mean(scores))
```

```python
accuracies
```

```python
from sklearn import metrics

for i in models:
    Y_pred=i.predict(X_test)
    matrix=metrics.confusion_matrix(Y_pred,Y_test)
    sns.heatmap(matrix,annot=True)
    plt.show()
```
